```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
func_lin = lambda x: [i for i in x ]
func_threshold = lambda x: [0 if i<0 else 1 for i in x ]
func_sigmoid= lambda x: [1/(1+math.e**(-i)) for i in x ]
func_tanh= lambda x: [(math.e**i-math.e**(-i))/(math.e**i+math.e**(-i)) for i in x ]
func_relu= lambda x: [0 if i<0 else i for i in x ]
func_lrelu= lambda x: [0.1*i if i<0 else i for i in x ]
func_elu= lambda x: [0.5*(math.e**i-1) if i<0 else i for i in x ]
func_softsign= lambda x: [i/(1+abs(i)) for i in x ]
```

```python
x=np.arange(-2,2,0.01)
plt.axhline(0) 
plt.axvline(0)
plt.grid()
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.plot(x,func_lin(x),c='r')
plt.savefig("images/func_lin.png")
```

```python
x=np.arange(-2,2,0.01)
plt.axhline(0) 
plt.axvline(0)
plt.grid()
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.plot(x,func_threshold(x),c='r')
plt.savefig("images/func_threshold.png")
```
